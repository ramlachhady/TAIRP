Data visualization in Bar plot

```python
dataset = pd.read_csv('annual_gold_rates.csv')

columnplot = ['USD', 'EUR', 'GBP', 'INR', 'AED', 'CNY']
plt.figure(figsize=(12, 6))
for col in columnplot:
    plt.bar(dataset['Date'], dataset[col], label=col)
plt.title('Annual Gold Rates by Currency over Years')
plt.xlabel('Year')
plt.ylabel('Gold Rate')
plt.legend()
plt.show()
```
