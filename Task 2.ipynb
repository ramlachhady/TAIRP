```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("IPL_Players_Data(2008-2022).csv") 
dataset
```

```python
dataset.head()
```

```python
print(dataset.isnull().sum())
dataset.dropna(inplace=True)
```

```python
numeric_columns = ['Year', 'Matches', 'Runss', 'Average', 'Strike Rate', 'NO', 'Fifty', 'Four', 'Six']
dataset[numeric_columns] = dataset[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
dataset['Average'] = dataset['Average'].astype(float)
```

```python
dataset['Higest_score'] = pd.to_numeric(dataset['Higest_score'], errors='coerce')
dataset['Century'] = dataset['Higest_score'].apply(lambda x: 1 if x >= 100 else 0)
```

```python
scaler = MinMaxScaler()
scaled_columns = ['Runss', 'Average', 'Strike Rate']
dataset[scaled_columns] = scaler.fit_transform(dataset[scaled_columns])
```

Visualization of player performance

```python
jos_buttler_data = dataset[dataset['Playerss'] == 'Virat Kohli']
plt.figure(figsize=(10, 6))
plt.plot(jos_buttler_data['Year'], jos_buttler_data['Runss'], marker='o')
plt.title("Jos Buttler's Runs Over Years")
plt.xlabel('Year')
plt.ylabel('Runs')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
metrics = ['Matches', 'Fifty', 'Four', 'Six']
values = jos_buttler_data[metrics].iloc[0]  
plt.bar(metrics, values)
plt.title("Jos Buttler's Performance Metrics")
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
jos_buttler_data = dataset[dataset['Playerss'] == 'Jos Buttler']
plt.figure(figsize=(10, 6))
plt.plot(jos_buttler_data['Year'], jos_buttler_data['Runss'], marker='o')
plt.title("Jos Buttler's Runs Over Years")
plt.xlabel('Year')
plt.ylabel('Runs')
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
metrics = ['Matches', 'Fifty', 'Four', 'Six']
values = jos_buttler_data[metrics].iloc[0] 
plt.bar(metrics, values)
plt.title("Jos Buttler's Performance Metrics")
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
plt.boxplot(jos_buttler_data['Runss'])
plt.title("Jos Buttler's Runs Distribution")
plt.ylabel('Runs')
plt.grid(True)
plt.show()
```
